# Feature Engineering

### Finding the NaN values

```python
df.nunique()
```

```python
df.isna().sum()
```

### Sleep Disorder 

filling the missing values in the sleep disorder feature

```python
df['sleep_disorder'].fillna('Normal', inplace=True)
```

### Gender - Label Encoding

```python
df = df.replace({'gender': {'Male': 1, 'Female' : 0}})
```

### BMI Category - Label Encoding

```python

df = df.replace({'bmi_category': {'Normal': 0, 'Normal Weight' : 1, 'Overweight' : 2, 'Obese' :3}})

df
```

### Blood Pressure

Removing unwanted values from blood pressure column

```python
df['blood_pressure'].unique()
```

```python
df.drop(df[df['blood_pressure'] == '_RARE_'].index, inplace=True)
```

Splitting the Blood Pressure Column to 'systolic' & 'diastolic'

```python
df[['systolic', 'diastolic']] = df['blood_pressure'].str.split('/', expand=True).astype(int)
```

```python
df.drop(columns=['blood_pressure'], inplace=True)
df
```

Exporting CSV 

```python
df.to_csv('dataset_preprocessed.csv', index=False)
```
